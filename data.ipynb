# Where do we find data? 

Whether observing at SBO or APO, you probably collected some neat data and want to play with it. Here's where you can find it.

## 24" Spectrograph Data from SBO 

- Data are automatically synced from the telescope control computer to `scorpius`. Once you log in, you can find all data for the semester at `/data/SBO-Data-Share/Classes/Spring-2026/ASTR3520`.
- Alternatively, you can use the direct link to the class OneDrive folder, which is shared on Canvas.

## KOSMOS Data from APO 

Whatever data you or your collaborators gathered while connected to TUI should already be downloaded onto your computer. However, if you worry you might be missing something or want additional calibration data, you can access the data in the following ways:

- We downloaded our entire S26 observing run to `scorpius`. You can find directories of data for every night, as well as guider images, on `scorpius` at `/home/zkbt/apo-astr3520-s26`.
- Follow [APO's instructions for downloading data](https://www.apo.nmsu.edu/mainpage/accessingdata/). You'll need a password that changes with every APO quarter; ask an instructor to get this password (and please, obviously, don't share it broadly).
