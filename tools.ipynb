# What tools might help? 

Here are a few resources you might find helpful at various stages in going from 2D raw pixels to 1D extracted spectra to scientific conclusions. This list is far from complete, and no matter what you'll likely need to write a good bit of analysis code yourself.

- [specsuite](https://github.com/autumn10677/specsuite), written by our very own Autumn Stephens, is a complete pipeline to go from 2D raw pixels to extracted spectra. It is particularly optimized for time-series observations unresolved point sources like stars, but many of its tools are more generally applicable. In particular, its clever warp modeling of how wavelengths curve across the detector pixels might also be useful for background-subtraction and/or defining extraction apertures for resolved objects like galacxies and nebula. *We (= Zach) strongly recommend using `specsuite` for as much as possible!*
- [specreduce](https://specreduce.readthedocs.io/en/latest/), an `astropy`-affiliated package, is a collection of tools for extracting 1D spectra from 2D images. It particular excels at defining flexible models for the trace (= the path a spectrum or region of a resolved astronomical object) takes across the detector. 
- [specutils](https://specutils.readthedocs.io/en/stable/), an `astropy`-affiliated pacakge, is useful for analyzing and fitting models to extracted 1D spectra. It has useful features for finding emission lines, fitting line profiles, estimating continua, and more!
- [pykosmos](https://github.com/jradavenport/pykosmos/), a package and set of notebooks for analyzing KOSMOS data or other longslit data. It doesn't have official documentation, but the `notebooks/` directory shows how it might be used.
